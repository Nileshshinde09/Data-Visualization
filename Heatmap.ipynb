```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('flights.csv')
```

```python
data.head()
```

```python
x = data.pivot_table(index = 'year',columns ='month',values = 'passengers',aggfunc= 'sum')
```

```python
sns.heatmap(x)
plt.plot()
```

```python
sns.heatmap(x,linewidth = 1,annot = True,fmt = 'd')
plt.plot()
```
