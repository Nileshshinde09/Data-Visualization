```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
sns.relplot(x = data.total_bill,y = data.tip,data = data)
plt.show()
```

```python
sns.relplot(x = data['total_bill'],y = data['tip'],hue = 'smoker',data = data)
plt.show()
```

```python
sns.relplot(x = data['total_bill'],y = data['tip'],hue = 'smoker',style = 'sex',size = 'size',data = data)
plt.show()
```

```python
sns.scatterplot(x = data['total_bill'],y = data['tip'],hue = 'smoker',data = data)
plt.show()
```
