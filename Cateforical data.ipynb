```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
sns.catplot(x = 'day',y = 'tip',kind = 'strip',data = data)
plt.show()
```

```python
sns.catplot(x = 'day',y = 'tip',kind = 'strip',jitter = 5,data = data)
plt.show()
```

```python
sns.catplot(x = 'day',y = 'tip',kind = 'swarm',data = data)
plt.show()
```

```python
sns.catplot(x = 'day',y = 'tip',kind = 'swarm',hue = 'sex',data = data)
plt.show()
```
