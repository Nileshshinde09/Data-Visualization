```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
sns.catplot(x= 'smoker',y ='tip',kind = 'bar',data = data)
plt.show()
```

```python
sns.catplot(x= 'smoker',y ='total_bill',kind = 'bar',data = data)
plt.show()
```

```python
sns.catplot(x= 'smoker',y ='total_bill',kind = 'bar',hue = 'sex',data = data)
plt.show()
```

```python
sns.catplot(x= 'smoker',y ='total_bill',kind = 'bar',estimator = np.mean,hue = 'sex',data = data)
plt.show()
```

```python
sns.catplot(x= 'sex',hue ='smoker',kind = 'count',data = data)
plt.show()
```
